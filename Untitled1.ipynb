```python
from random import gauss
from random import seed
from pandas import Series
from pandas.plotting import autocorrelation_plot
import numpy as np
# seed random number generator

Samples = []


for j in range(500):
  seed(j)
  inner_list = []
  # create white noise series
  series = [gauss(0.0, 1.0) for i in range(1000)]
  inner_list = series
  Samples.append(inner_list)
```

```python
granted = []

for m in range(500):
  series = []
  for p in range (1000):
    if p<500:
      series.append(1)
    else:
      series.append(0)
  granted.append(series)
```

```python
from sklearn.model_selection import train_test_split



Samples = np.array(Samples, dtype="object")
granted = np.array(granted, dtype="object")

print(Samples.shape)
print(granted.shape)

X_train, X_test,y_train,y_test = train_test_split(Samples,granted, test_size=0.33, 
                                                    )
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)

```

```python
#potem wybudowanie sieci czyli stworzenie ich warstw(jej wygląd zależy ode mnie)
import tensorflow as tf
from tensorflow import keras 

model = keras.Sequential(
    [
      keras.layers.Input(shape=(1000,)),
      keras.layers.Dense(1000,activation="ReLU"),
      keras.layers.Dense(1024,activation="ReLU"),
      keras.layers.Dense(2048,activation="ReLU"),
      keras.layers.Dense(1998,activation="ReLU",name="layer1"),
      keras.layers.Dense(1000,activation="sigmoid"),
    ]
)

outputs=[layer.output for layer in model.layers]

print(outputs[3])

print(X_train.shape)
print(y_train.shape)

X_train = np.asarray(X_train).astype(np.float32)
X_test = np.asarray(X_test).astype(np.float32)
y_train = np.asarray(y_train).astype(np.float32)
y_test = np.asarray(y_test).astype(np.float32)

```

```python

import tensorflow as tf
from scipy import signal
from keras.layers import Dense, Flatten
from keras import backend as K

tf.config.run_functions_eagerly(False)



def my_loss(outputs,granted):
  f2, psd2 = signal.welch(outputs[3],nperseg=1000)
  f, psd = signal.welch(granted,nperseg=1000)
  return keras.losses.MSE(psd, psd2)

model.summary()
layer_name = 'layer1'
intermediate_layer_model = model(inputs=model.input,
                                 outputs=model.get_layer(layer_name).output)
intermediate_output = intermediate_layer_model.predict(granted)
model.compile(optimizer='adam', loss=my_loss(outputs[3],granted), run_eagerly=True)


history = model.fit(X_train,y_train,epochs=512,batch_size=10)
```
